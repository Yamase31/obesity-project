```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Read in filename and set the index: df
rows = ['Korea', 'United States', 'United Kingdom', 'Mexico', 'Japan']
cols = ['Value', 'Time', 'Country']
df = pd.read_csv('gdp_cap.csv')
sns.set()
df = df[df.Country.isin(rows)]
df = df.loc[:,cols]
df = df.pivot(index = 'Time',columns = 'Country',values = 'Value')
display(df)

#plot df
df.plot(figsize = (10,8))
plt.legend(prop={'size':12}, loc='best')
plt.xlabel('Year', fontsize=14)
plt.ylabel('GDP per Capita (USD)', fontsize=14)
plt.title('GDP per Capita in OECD Countries', fontsize=16)
plt.show()
```

```python
#set df to specific time frame
df = df[df.index>=2000]
df = df[df.index<2018]

#display(df)
df1=pd.read_csv('C:/dfTotalPopulation.csv')
df1=df1.dropna(how='any')
sns.set()

df1_reg=df1[df1.Country.isin(rows)].pivot(index= 'Year', columns='Country', values='Value')
df1_reg.head(20)
display(df1_reg)
df[df1_reg.iloc[:,[1]].notna().iloc[:,[0]]].iloc[:,[1]].dropna(how = 'any')

#plot df with for loop and scatter function
plt.figure(figsize = (10,8))
for index in range(5):
    df_temp = df[df1_reg.iloc[:,[index]].notna().iloc[:,[0]]].iloc[:,[index]].dropna(how = 'any')
    df1_temp =  df1_reg[df1_reg.iloc[:,[index]].notna().iloc[:,[0]]].iloc[:,[index]].dropna(how = 'any')
    plt.scatter(df1_temp.index,df1_temp.iloc[:,[0]], label = df_temp.columns.astype('str')[0], s = df_temp.iloc[:,[0]]/300)
plt.xlabel('Year', fontsize=14)
plt.ylabel('Obese Population (%)', fontsize=14)
plt.title('Correlation of Obesity and GDP per Capita in OECD Countries, size = GDP per capita', fontsize=16)
plt.legend()
plt.savefig('obese_gdp_per_cap.jpg')
plt.legend()
```
